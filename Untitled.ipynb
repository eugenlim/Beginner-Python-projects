# 5 - Cropping 

### We can crop specific region of image by method crop. This crop method accepts a 4 - valued tuple represented as (left, top, right, bottom). Where left, top, right and bottom represents the pixel coords of area you want to crop out. It returns an image object

```python
left = 50
top = 120
right = 250
bottom = 230 
crop_image = image.crop((left, top, right, bottom))
plt.imshow(crop_image)
```

# 7 - Transposing
### We can change the orientation of image using transpose function. Transpose function takes one parameter that is method of transpose. There are six transposing method you can use. They are as follows:

#### Image.FLIP_LEFT_RIGHT(This inverse the row) 
#### Image.FLIP_TOP_BOTTOM(This inverse the columns)
#### Image.ROTATE_90 (This rotates the image by an angle of 90)
#### Image.ROTATE_180 (This rotates the image by an angle of 180)
#### Image.ROTATE_270 (This rotates the image by an angle of 270)
#### Image.TRANSPOSE (This converts the rows into columns)
#### This returns the image object


```python
plt.figure(figsize=(10,10))
transpose_image1=image.transpose(Image.FLIP_LEFT_RIGHT)
plt.subplot(3,2,1)
plt.imshow(transpose_image1)
plt.title("FLIP_LEFT_RIGHT")

transpose_image2=image.transpose(Image.FLIP_TOP_BOTTOM)
plt.subplot(3,2,2)
plt.imshow(transpose_image2)
plt.title("FLIP_TOP_BOTTOM")

transpose_image3=image.transpose(Image.ROTATE_90)
plt.subplot(3,2,3)
plt.imshow(transpose_image3)
plt.title("ROTATE_90")

transpose_image4=image.transpose(Image.ROTATE_180)
plt.subplot(3,2,4)
plt.imshow(transpose_image4)
plt.title("ROTATE_180")

transpose_image5=image.transpose(Image.ROTATE_270)
plt.subplot(3,2,5)
plt.imshow(transpose_image5)
plt.title("ROTATE_270")

transpose_image6=image.transpose(Image.TRANSPOSE)
plt.subplot(3,2,6)
plt.imshow(transpose_image6)
plt.title("TRANSPOSE")
```
